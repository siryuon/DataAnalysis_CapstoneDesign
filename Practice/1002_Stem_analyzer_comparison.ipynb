```python
import pandas as pd
train_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/train_data.csv', encoding='cp949')
train_data['comments'] = train_data['comments'].str.replace('[^ㄱ-ㅎ ㅏ-ㅣ 가-힣]','')
```

```python
samples = train_data['comments'][:50]
```

```python
from konlpy.tag import Hannanum, Kkma, Komoran, Mecab, Okt
import time
analyzer_list = [('hannanum', Hannanum()), ('kkma', Kkma()), ('Komoran', Komoran()), ('mecab', Mecab()), ('Okt', Okt())]
results = []
for name, analyze in analyzer_list:
    stems = []
    process_time = time.time()
    
    for sample in samples:
        stems.append(analyze.morphs(sample))
    process_time = time.time() - process_time
    print('Analyzer name: %10s, %.3f secs' % (name, process_time))
    results.append((name,stems))
```

```python
result_df = pd.DataFrame(results, columns=['형태소 분석기','내용'])
result_df[:10]
```
